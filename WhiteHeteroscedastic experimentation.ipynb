```python
import numpy as np
import GPy
import matplotlib.pyplot as plt
%matplotlib inline

X = np.arange(0,10,1)[:,None]
y = np.ones(10)[:,None]
y[3]+=2
```

```python
var = np.ones(10)*0.01
#var[3]=10
k = GPy.kern.RBF(1,lengthscale=2)+GPy.kern.WhiteHeteroscedastic(1, 10, variance=var)
m = GPy.models.GPRegression(X,y,k)
m.Gaussian_noise.variance.fix(0.1)
```

```python
m.plot()
plt.errorbar(X,y,np.sqrt(var),fmt = '', color='k', linestyle='None')
```
