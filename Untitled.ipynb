```python

A = np.array([[4,1,0],[1,2,0],[0,0,2]])
B = np.array([[3,0,0],[0,3,0],[0,0,3]])
ev, P = np.linalg.eig(A)
Pinv = np.linalg.inv(P)
```

```python
def expand_covariance(A,B):
    """Expand A to include all of B's covariances"""
    newA = A.copy()
    print "==="
    while (True): #this will end when there are no eigenvalues>1
        print "---"
        T = np.dot(np.linalg.inv(newA),B)
        print "T:"
        print T
        evals,P = np.linalg.eig(T)
        indices = np.where(evals>1.0)[0]
        print "eigenvalues"
        print evals
        if len(indices)<1:
            return newA #nothing to do
        index = indices[0]
        print "index: %d" % index
        Pinv = np.linalg.inv(P)
        Atwist = np.dot(np.dot(Pinv,newA),P)
        print "Atwist"
        print Atwist
        S = np.eye(len(A))
        S[index,index] = evals[index]
        newA = np.dot(S,Atwist)
        print "S.Atwist"
        print newA
        newA = np.dot(np.dot(P,newA),Pinv)
        print "rotated back"
        print newA
```

```python
def extend_cov(A,B):
    """Extend A to include B"""
    Anew = A.copy()
    for i in range(4):
        #calculate the square roots of A and B using eigenvalue decomposition
        La,Va = np.linalg.eig(Anew)
        La = np.diag(La)
        Lb,Vb = np.linalg.eig(B)
        Lb = np.diag(Lb)
        Ta = np.dot(Va,np.sqrt(La))
        np.dot(Ta,Ta.transpose())
        Tb = np.dot(Vb,np.sqrt(Lb))
        np.dot(Tb,Tb.transpose())

        #Ta Ta^T is the covariance of A, so Ta is the square root of that.
        #here we invert this, as we want to apply the opposite transformation
        #(i.e. we want to rotate and scale in the opposite direction)
        Ta = np.linalg.inv(Ta)
        #this is the transformed covariance B, by the inverse of A
        Z = np.dot(Ta,np.dot(Tb,np.dot(Tb.transpose(),Ta.transpose())))
        #we get the eigenvalues and vectors
        L,V = np.linalg.eig(Z)

        #build new scaling matrix
        scalemat = np.eye(len(A))
        
        indices = np.where(L>1.0)[0]
        if len(indices)<1:
            print Anew
            print "Z"
            print Z
            return Anew,Z #nothing to do
        index = indices[0]
        print L,index
        scalemat[index,index] = L[index]
        print "Scale mat"
        print scalemat
        Tfix = np.dot(V.transpose(),np.dot(np.sqrt(scalemat),V))
        Anew = np.dot(np.dot(Tfix,A),Tfix.transpose()) 
        print Anew
        
```
